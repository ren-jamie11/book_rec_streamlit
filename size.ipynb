```python
import pandas as pd
import os

file_paths = [
    "data/all_books_final.parquet",
    "data/users_data.parquet",
    "data/genre_labels.parquet",
    "data/all_labeled_reviews.parquet",
    "data/compact_user_genre_pct.parquet",
    "data/smaller_user_item_matrix.parquet"
]

total = 0
for path in file_paths:
    if os.path.exists(path):
        df = pd.read_parquet(path)
        size = df.memory_usage(deep=True).sum() / 1e6
        print(f"{os.path.basename(path)}: {size:.2f} MB")
        total += size
    else:
        print(f"{path} not found.")

print(f"\nTotal memory usage: {total:.2f} MB")
```
